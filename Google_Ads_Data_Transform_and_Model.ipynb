```python
import re

# function to extract numeric values from a string using regular expressions
def clean_price(string):
    # use regular expressions to match numeric values
    match = re.findall(r'\d+', string)
    # join the matched values to create a string of numeric characters
    return '-'.join(match)

# create a new column 'cleaned_price' with the cleaned price values
df['cleaned_price'] = df['budget'].apply(clean_price)
```

```python
df['cleaned_price'].unique()
```

```python
df['price_frequency'] = ''

# loop through the 'price' column and extract the frequency of the price
for index, row in df.iterrows():
    # use regular expressions to match the price frequency indicator
    if re.search(r'pw\b', row['budget']):
        df.at[index, 'price_frequency'] = 'weekly'
    elif re.search(r'pm\b', row['budget']):
        df.at[index, 'price_frequency'] = 'monthly'
    elif re.search(r'py\b', row['budget']):
        df.at[index, 'price_frequency'] = 'yearly'
    elif 'week' in row['budget']:
        df.at[index, 'price_frequency'] = 'weekly'
    elif 'month' in row['budget']:
        df.at[index, 'price_frequency'] = 'monthly'
    elif 'year' in row['budget']:
        df.at[index, 'price_frequency'] = 'yearly'

# print the updated dataframe
df.head()
```

```python
def duration_to_weeks(duration_str):
    duration_str = duration_str.lower()

    # convert common formats to weeks
    if 'week' in duration_str:
        num_weeks = re.findall(r'\d+', duration_str)
        if len(num_weeks) > 0:
            return int(num_weeks[0])
    elif 'month' in duration_str:
        num_months = re.findall(r'\d+', duration_str)
        if len(num_months) > 0:
            return int(num_months[0]) * 4
    elif 'year' in duration_str:
        num_years = re.findall(r'\d+', duration_str)
        if len(num_years) > 0:
            return int(num_years[0]) * 52

    # return -1 if duration is not in any of the above formats
    return -1
    df['duration_weeks'] = df['lease'].apply(duration_to_weeks)
```

```python
def remove_non_numeric(column):
    column = column.apply(lambda x: re.sub(r'[^\d-]', '', str(x)))
    return column

df['lease'] = remove_non_numeric(df['lease'])
```
